```python
import seaborn as sns

flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights_pivot = flights.pivot(index='month',
                             columns='year',
                             values='passengers')
flights_pivot
```

```python
sns.heatmap(data=flights_pivot)
```

```python
sns.lineplot(x='year', y='passengers', data=flights)
```
